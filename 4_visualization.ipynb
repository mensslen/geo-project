```python
import json
import pandas as pd
```

```python
df = pd.read_csv("random_zip_codes_with_coordinates_distances_and_radius.csv")
```

```python
df
```

```python
# We define a simple color range.
def assign_color(x):
    if x < 3:
        color = "#ffc1bd"
    elif x < 10:
        color = "#ff6157"
    elif x >= 10:
        color = "#fc0f00"
    return color

df["color"] = df["quantity"].apply(assign_color)
```

```python
df
```

```python
# We build a geojson structure with dictionaries.
data = {}
locations = []
for i in df.itertuples():
    row = {
        "geometry": 
        {"coordinates": [float(i[3]), float(i[4])], "type": "Point"}, 
        "properties": 
        {"zip code": i[1], 
         "quantity": i[2],
         "marker-size": "small",
         "marker-color": i[-1]}, 
         "type": "Feature"
    }
    locations.append(row)
        
#"marker-symbol": i.frequency},
```

```python
data["features"] = locations
data["type"] = "FeatureCollection"
# It should work with the common json module.
json_file = json.dumps(data)
with open("geo_data_of_users.json", "w") as f:
    f.write(json_file)
# It works, but we might have to open the json file in geojson.io manually if it's too big.
```
