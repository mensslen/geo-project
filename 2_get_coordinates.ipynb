```python
import pandas as pd
from pprint import pprint
```

```python
df = pd.read_csv("random_zip_codes.csv")
```

```python
# Get quantity of duplicates via pivot table.
duplicates = df.pivot_table(index=["zip"], aggfunc='size')
```

```python
# Turn duplicates into a new data frame.
df = duplicates.reset_index()
```

```python
df.rename(columns={0:'quantity'}, inplace=True) 
```

```python
df.to_csv("aggregated_locations_of_users.csv", index=False)
```

```python
# Open a aggregated set of random zip codes.
df = pd.read_csv("aggregated_locations_of_users.csv", converters={"zip": lambda x: str(x)})
```

```python
# Open prepared registry of zip codes.
df2 = pd.read_csv("prepared_registry_german_zip_codes.csv", converters={'postal code': lambda x: str(x)}, )
```

```python
df2
```

```python
# Function to match aggregated set of random zip codes with prepared zip code registry. Returns mean coordinates.
def check_file(x):
    mask = df2["postal code"].str.contains(x)
    # Sometimes there are several loactions associated with one zip. We calculate mean coordinates.
    longitude = df2[mask]["longitude"].mean()
    latitude = df2[mask]["latitude"].mean()
    return longitude, latitude
```

```python
# Assigns coordinates as new columns to set of random zip codes.
df["av. longitude"], df["av. latitude"] = zip(*df["zip"].apply(check_file))
```
